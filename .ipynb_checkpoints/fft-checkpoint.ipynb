```python
import numpy as np
import matplotlib.pyplot as plt

# ----- signal -----
Fs = 1000
N = 1000
t = np.linspace(0, 1, N, endpoint=False)

# clean signal
sig_clean = np.sin(2*np.pi*300*t) + 0.5*np.sin(2*np.pi*120*t)

# add Gaussian noise
noise_std = 3.0              # try 0.2, 0.5, 1.0, 2.0
noise = np.random.normal(0, noise_std, size=N)

sig = sig_clean + noise

# ----- FFT -----
X = np.fft.rfft(sig)
freqs = np.fft.rfftfreq(N, d=1/Fs)
magnitude = np.abs(X)

# ----- plot -----
plt.plot(freqs, magnitude)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude |FFT|")
plt.title("Frequency Spectrum with Gaussian Noise")
plt.grid(True)
plt.show()

print(sig_clean)
```
